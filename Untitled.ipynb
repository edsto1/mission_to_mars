```python
from splinter import Browser
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://mars.nasa.gov/news/?page=1&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

news = soup.find_all('li', class_='slide')

for story in news:
    news_title = story.find('div',class_='content_title').text
    news_p = story.find("div", class_="article_teaser_body").text
    print("Title: " + news_title)
    print()
    print("Summary: " + news_p)
    print('==========================')
        
try:
    browser.click_link_by_partial_text('MORE')
          
except:
    print("Scraping Complete")
```
